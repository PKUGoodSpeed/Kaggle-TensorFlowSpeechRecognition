```python
train_dir = '/Users/goodspeed/Documents/Kaggles/Kaggle-TensorFlowSpeechRecognition/pkugoodspeed/data/train/audio'
```

```python
if os.path.exists(train_dir + '/' + '_background_noise_'):
    os.system('mv {0}/_background_noise_ {1}/silence'.format(train_dir, train_dir))
```

```python
labels = os.listdir(train_dir)
print(labels)
print(len(labels))
```

```python
if os.path.exists(train_dir+'/silence/README.md'):
    os.system('rm {0}/silence/README.md'.format(train_dir))
```

```python
raw = {'x': [], 'y': [], 'label':[]}
for i, folder in enumerate(labels):
    print(folder)
    for file in os.listdir(train_dir + '/' + folder):
        rate, samples = wavfile.read(train_dir + '/' + folder + '/' + file)
        assert(rate == 16000)
        raw['x'].append(np.array(samples))
        raw['y'].append(i)
        raw['label'].append(folder)
```

```python
df = pd.DataFrame(raw)
display(df)
```

```python
df.to_csv('../data/raw.csv')
```
